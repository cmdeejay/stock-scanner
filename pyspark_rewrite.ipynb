```python
import pandas as pd
from pyspark.sql import SparkSession
import yfinance as yf
from datetime import date
from dateutil.relativedelta import relativedelta
from pandas import ExcelWriter
import numpy as np
```

```python
spark = SparkSession.builder.appName('report').getOrCreate()
```

```python
## Tickers
filepath = r"files/weeklyreport.csv"
df = spark.read.csv(path=filepath, header=True)
tickers = df.select('股票代码').rdd.flatMap(lambda x: x).collect()
```

```python
## Transform tickets

new_tickers = []
for ticker in tickers:
    if int(ticker) > 600000:
        ticker = ticker + ".SS"
    else:
        ticker = ticker + ".SZ"
    new_tickers.append(ticker)

new_tickers
```

```python
today = date.today()
start = today - relativedelta(months=1)
data = yf.download(tickers=new_tickers[1:5], start=start, end=today)
data = data["Adj Close"]
```
